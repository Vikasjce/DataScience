Sampling Methods 
1. Simple random Sampling - each case is equally likely to be selected eg: randomly pulling names from the hat 

2. Stratified sampling
divide the population into homogenous groups and then randomly sample from the group 
eg: males and female group and then randomly sample from the group 

3. cluster Sampling
1. divide the population into heterogenous clusters 
2. randomly select the clusters 
3. then sample all observations from these clusters 

4. Multistage Sampling
1. divide the population into heterogenous clusters 
2. randomly select the clusters 
3. then randomly sample observations from these clusters

3. Mode
  -- repeatetive numbers
